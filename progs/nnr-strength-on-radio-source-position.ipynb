This notebook shows the NNR strength on the radio source position estimates in the VLBI solutions.

```python
from astropy.table import Table, Column
from astropy.time import Time
import astropy.units as u
import bottleneck as bn
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
from my_progs.vlbi.read_sou import read_sou
from my_progs.catalog.pos_diff import radio_cat_diff_calc

free_sou = read_sou("../data/gcrf-a1u/gcrf-a1u.sou")
fix_nnr20_sou = read_sou("../data/gcrf-a1u-fix/gcrf-a1u-fix.sou")

# Calculate the positional difference
souoft1 = radio_cat_diff_calc(fix_nnr20_sou, free_sou, sou_name="iers_name")
```

```python
fig, (ax0, ax1) = plt.subplots(figsize=(8, 8), nrows=2, sharex=True)

ax0.plot(souoft1["ra"], souoft1["dra"], ".", ms=1)
ax1.plot(souoft1["ra"], souoft1["ddec"], ".", ms=1)

ax0.set_ylabel("Offset in R.A. [mas]")

ax1.set_xlabel("Right ascension [$^\circ$]")
ax1.set_ylabel("Offset in decl. [mas]")

ax1.set_xticks(np.arange(0, 361, 60))

ax0.set_xlim([0, 360])
ax0.set_ylim([-1, 1])
ax1.set_ylim([-1, 1])

ax0.set_title("When NNR = 20 uas", fontsize=15)

ax0.xaxis.set_ticks_position("both")
ax0.yaxis.set_ticks_position("both")
ax1.xaxis.set_ticks_position("both")
ax1.yaxis.set_ticks_position("both")

plt.subplots_adjust(hspace=0.1)
```

```python
fig, (ax0, ax1) = plt.subplots(figsize=(8, 8), nrows=2, sharex=True)

ax0.plot(souoft1["dec"], souoft1["dra"], ".", ms=1)
ax1.plot(souoft1["dec"], souoft1["ddec"], ".", ms=1)

ax0.set_ylabel("Offset in R.A. [mas]")

ax1.set_xlabel("Declination [$^\circ$]")
ax1.set_ylabel("Offset in decl. [mas]")

ax1.set_xticks(np.arange(-90, 91, 30))

ax0.set_xlim([-90, 90])
ax0.set_ylim([-1, 1])
ax1.set_ylim([-1, 1])

ax0.set_title("When NNR = 20 uas", fontsize=15)

ax0.xaxis.set_ticks_position("both")
ax0.yaxis.set_ticks_position("both")
ax1.xaxis.set_ticks_position("both")
ax1.yaxis.set_ticks_position("both")

plt.subplots_adjust(hspace=0.1)
# plt.savefig("../plots/gcrf250fix-sou-diff-on-dec1.png")
```
