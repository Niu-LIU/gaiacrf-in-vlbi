# Check the galactic aberration effect in the VLBI solution (2)

To check the Galactic aberration effect, I run two test solutions with or without the GA modeling and compare these two solutions.

This comparison is perfomred within the frame of ICRF2 and ICRF3, respectively.

Here I present the one of ICRF3.

In this notebook I will look at the radio source catalogs (.sou file) from a VLBI solution 
denoted as `icrf3-ga-19` and `icrf3-nga-19`.

The comparison carried out here is similar to that in the notebook `check-ga-on-icrf2` 
but all VLBI sessions up to 2019 are used.

Only the radio source positions are compared.

Three comparisons need to do done:

- `icrf3-nga-19` vs. `icrf3-ga-19`
- `icrf3-nga-19` &`icrf3-ga-19` vs. `ICRF3`
- `icrf3-nga-19` &`icrf3-ga-19` vs. `Gaia DR2`

The first comparison is to check if I can nearly reproduce the ICRF2 solution.

```python
from astropy.table import Table, Column
import astropy.units as u
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

import bottleneck as bn
```

## 1 `icrf3-nga-19` vs. `icrf3-ga-19`

This result is similar to that when using the ICRF2 catalog to provide the a priori source position.

```python
# My modules
from my_progs.catalog.vsh_deg2_cor import vsh_deg02_fitting, residual_calc02
from my_progs.catalog.write_output import print_vsh1_corr, print_vsh2_corr
```

```python
# Try with all the sources
# Transform columns into np.array
dra = np.array(soldif["dra"])
ddec = np.array(soldif["ddec"])
dra_err = np.array(soldif["dra_err"])
ddec_err = np.array(soldif["ddec_err"])
ra_rad = np.array(soldif["ra"].to(u.radian))
dec_rad = np.array(soldif["dec"].to(u.radian))
dra_ddec_cov = np.array(soldif["dra_ddec_cov"])

# Transformation parameters
# l_max = 2
w2_all, sig2_all, corrcoef2_all, _, _, _ = vsh_deg02_fitting(
    dra, ddec, ra_rad, dec_rad, dra_err, ddec_err,
    cov=dra_ddec_cov, elim_flag="None")

# mas -> uas
w2 = w2_all * 1.e3
sig2 = sig2_all * 1.e3

# Print results
print("Estimates (%6d sources)\n"
      "----------------------------------------------"
      "----------------------------------------------\n"
      "              Rotation [uas]                  "
      "                 Glide [uas]               \n"
      "     x             y             z           "
      "     x             y             z\n"
      "----------------------------------------------"
      "----------------------------------------------\n"
      "%+4.0f +/- %3.0f  %+4.0f +/- %3.0f  %+4.0f +/- %3.0f  "
      "  %+4.0f +/- %3.0f  %+4.0f +/- %3.0f  %+4.0f +/- %3.0f\n"
      "----------------------------------------------"
      "----------------------------------------------\n" %
      (dra.size, 
       w2[3], sig2[3], w2[4], sig2[4], w2[5], sig2[5],
       w2[0], sig2[0], w2[1], sig2[1], w2[2], sig2[2]))

quad_names = Column(["ER22", "EI22", "ER21", "EI21", "E20",
                     "MR22", "MI22", "MR21", "MI21", "M20"])
t_quad = Table([quad_names, w2[6:], sig2[6:]], 
               names=["Quadrupolar term", "Estimate", "Error"])
t_quad["Estimate"].format = "%5.0f"
t_quad["Error"].format = "%5.0f"
print(t_quad)


print("Correlation coefficient between parameters in 'l_max=2' fit")
print_vsh2_corr(corrcoef2_all, deci_digit=1, included_one=False)
```

## 2 Comparison with the ICRF3 S/X catalog

This comparison shows very similar results to that of using the ICRF2 catalogs as the a priori catalog.

One interesting finding is that the VSH parameters modelled from the positional offsets between the 
ICRF3 catalog and VLBI solutions using the ICRF3 catalog as the apriori catalogs, are almost the same
with that from the positional offset between ICRF3 catalog and solutions using the ICRF2 catalog as the 
a priori catalog, except the three rotation terms.

The `icrf2` solutions show insignificant rotation terms with respect to the ICRF3 catalog while the `icrf3`
solutions yiled rotation terms of $R_x$ ~ 25 $\mu$as and $R_y$ ~ 50 $\mu$as, although these terms are very 
small compared to their formal uncertainty.

In my opinion, this is due to the fact that the ICRF3 catalog results from a VLBI solution using the ICRF2
catalog as the a priori catalog.

```python
from my_progs.catalog.read_icrf import read_icrf3
icrf3 = read_icrf3()

# Crossmatch between ICRF3 and solution
icrfdif1 = radio_cat_diff_calc(solnga, icrf3, sou_name="iers_name")
icrfdif2 = radio_cat_diff_calc(solga, icrf3, sou_name="iers_name")

icrfdif1.sort("dec")
icrfdif2.sort("dec")
```
